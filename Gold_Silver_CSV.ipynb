```python
# importing the libraries and creating a data dictionary 

from bs4 import BeautifulSoup
from collections import defaultdict
import requests

# data to post
data = {"action": "historical_data",
       "curr_id": "8830",
       "st_date": "02/01/2017",
       "end_date": "03/06/2018",
       "interval_sec": "Daily"}

# add a user agent and specify that we are making an ajax request
head = {

       "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
       "X-Requested-With": "XMLHttpRequest"}

# creating Ajax call to fetch the historical data of date and price columns only

with requests.Session() as s:
   url='https://www.investing.com/instruments/HistoricalDataAjax'
   r = s.post(url, data=data, headers=head)
   od = defaultdict()
   soup = BeautifulSoup(r.content, "lxml")

   table = soup.select_one("table.genTbl.closedTbl.historicalTbl")
   cols = [th.text for th in table.select("th")][0:2]
   for row in table.select("tr + tr"):
       data = [td.text for td in row.select("td")]
       od[data[0]] = dict(zip(cols, data[0:2]))

from pprint import pprint as pp

pp(dict(od))

# writing to a CSV file

import csv
with open('gold.csv', 'w') as f:
   csvwriter = csv.writer(f, delimiter=';', quoting=csv.QUOTE_NONE)
   csvwriter.writerow(cols)
   for session in od:
       for item in od[session]:
           if item.upper() == "PRICE":
               csvwriter.writerow([session,od[session][item]])
```
